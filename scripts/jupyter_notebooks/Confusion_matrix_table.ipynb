# Obtaining table
here I will extract the opt-1.txt and opt-2.txt files and I will handle them as pandas tables.

```python
import pandas as pd
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
def title_extracter(filename):
  col_names = []
  with open(filename, 'r') as fh:
    line = fh.readline()
    wd = line.split()
    for i in range(0,len(wd),2):
      col_names.append(wd[i][:-1])
  return col_names



```

```python
col_names = title_extracter('opt-table-1.txt')
```

### Starting with opt-table-1.txt

```python
table_1 = pd.read_csv('opt-table-1.txt', sep=' ', header=None).iloc[:,1:14:2]
```

```python
table_1.head()
```

```python
or_names = [i for i in range(1,15,2)]
```

```python
table_1 = table_1.rename({i:j for i,j in zip(or_names,col_names)}, axis=1) #renaming using the original:new_name mapping
```

```python
table_1
```

```python
table_1.to_excel('opt-table1.xlsx', index=False)
```
