### Adding the concatenated column

```python
df2['id:chain'] = df2['PDB ID'].str.strip()+":"+df2['PDB chain ID'].str.strip()
```

Now we print it

```python
def chain_printer(df):
  for index,row in df.iterrows():
    print(row['id:chain'])
```
